## Verificando a diabetes pelo fator Gênero

```python
# Verificando gênero para quem possui diabetes:

labels=['Male','Female']
plt.title('Diabetes por gênero (%)')
plt.pie(Dados_diabetes['gender'].value_counts(), labels = labels, autopct='%.02f');
plt.annotate(
    'Análise sobre Diabetes \n by: @Lerina Mesquita Mastruian',
    xy=(1, -0.20), xycoords='axes fraction', ha='right', fontsize=12);
plt.show()
```

Comparativo Hipertensão e Doenças Cardiacas para quem possui Diabetes

```python
#Criação de uma tabela de contingência (crosstab) para comparar a presença de hipertensão e doença cardíaca em pessoas com e sem diabetes:

ct = pd.crosstab(Dados["diabetes"], [Dados["hypertension"], Dados["heart_disease"]], margins=True)
print(ct)
```

```python
#Criação de um gráfico de barras empilhadas para visualizar a distribuição da Hipertensão e doença cardíaca em pessoas com e sem diabetes:

sns.catplot(data=Dados, kind="count", x="diabetes", hue="hypertension", col="heart_disease")
plt.title('Verificação de relação da Diabetes com Hipertensão e Doenças do Coração')
plt.annotate(
    'Análise sobre Diabetes \n by: @Lerina Mesquita Mastruian',
    xy=(1, -0.30), xycoords='axes fraction', ha='right', fontsize=12);
plt.show()
```

Na verificação se haveria uma relação se existe uma relação entre tabagismo e diabetes, no grafico acima, mostra que temos um volume maior de pessoas que não fumam com histórico de tabagismo e nao possuem diabetes, dessa forma, não temos relação entre essas variáveis

## Calculo da porcentagem de pessoas com hipertensão e doença cardíaca em cada grupo (com diabetes e sem diabetes):

```python
Dados_diabetes = Dados[Dados["diabetes"] == 1]
Dados_no_diabetes = Dados[Dados["diabetes"] == 0]

pct_hipertensao_diabetes = Dados_diabetes["hypertension"].sum() / len(Dados_diabetes) * 100
pct_hipertensao_no_diabetes = Dados_no_diabetes["hypertension"].sum() / len(Dados_no_diabetes) * 100

pct_heart_disease_diabetes = Dados_diabetes["heart_disease"].sum() / len(Dados_diabetes) * 100
pct_heart_disease_no_diabetes = Dados_no_diabetes["heart_disease"].sum() / len(Dados_no_diabetes) * 100

print("Porcentagem de pessoas com hipertensão e diabetes: {:.2f}%".format(pct_hipertensao_diabetes))
print("Porcentagem de pessoas com hipertensão e sem diabetes: {:.2f}%".format(pct_hipertensao_no_diabetes))
print("Porcentagem de pessoas com doença cardíaca e diabetes: {:.2f}%".format(pct_heart_disease_diabetes))
print("Porcentagem de pessoas com doença cardíaca e sem diabetes: {:.2f}%".format(pct_heart_disease_no_diabetes))
```

Podemos concluir que, os individuos que possuem diabetes possuem maior propensão de ter doenças cardiacas e hipertensão do que as pessoas que nao possuem a diabetes.

Distribuição do IMC por grupo

```python
Dados_diabetes = Dados[Dados["diabetes"] == 1]
Dados_no_diabetes = Dados[Dados["diabetes"] == 0]

sns.histplot(data=Dados_diabetes, x="bmi", kde=True, color="red")
sns.histplot(data=Dados_no_diabetes, x="bmi", kde=True, color="blue")
plt.title('Distribuição do IMC para os grupos: Diabéticos x Não Diabéticos')
plt.annotate(
    'Análise sobre Diabetes \n by: @Lerina Mesquita Mastruian',
    xy=(1, -0.20), xycoords='axes fraction', ha='right', fontsize=8);
plt.show()
```

## Calculo do IMC para cada grupo

```python
media_bmi_diabetes = Dados_diabetes["bmi"].mean()
media_bmi_no_diabetes = Dados_no_diabetes["bmi"].mean()

print("A média do IMC em pessoas com diabetes: {:.2f}".format(media_bmi_diabetes))
print("A média do IMC em pessoas sem diabetes: {:.2f}".format(media_bmi_no_diabetes))
```

# Perfil de risco para diabetes

```python
#Vamos identificar o perfil de risco para o desenvolvimento do diabetes com base nos fatores de risco
#Dessa forma vamos identificar quais são os fatores que estariam mais associados ao desenvolvimento do diabetes

Dados = Dados[["age", "gender", "bmi", "smoking_history", "hypertension", "heart_disease", "diabetes"]]
```

```python
#vamos converter as variaveis categóricas em variáveis numericas usando configuração one-hot (dummy variables)

Dados = pd.get_dummies(Dados, columns=["gender", "smoking_history"],drop_first=True)
```

```python
#Calcular a matriz de correlação para identificar onde podemos associar os fatores de risco e variavel "Diabetes"

correlation_matrix = Dados.corr()
print(correlation_matrix)
```

```python
#Vamos visualizar a matriz através de um mapa de calor

sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.show()
```

Avaliando os resultados da matriz de calor, avaliando a questão de Diabetes, quais seriam os fatores com forte correlação, poderiamos elencar em ordem de importância, seriam : Idade, Índice de massa corpórea, Hypertensão, e Doenças do coração.
